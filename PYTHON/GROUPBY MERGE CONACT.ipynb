```python
import pandas as pd
import numpy as np
```

```python
Dict1={
    'Column1':['x','y','z','x','y','y'],
    'Column2':[5,3,2,1,4,7],
    'Column3':[10,20,30,40,50,30],
    'Column4':['c','a','b','a','c','b']
}
```

```python
Data1=pd.DataFrame(Dict1)
```

```python
Data1
```

```python
Grp=Data1.groupby('Column1')
Grp
```

```python
for i,Data in Grp:
    print(i)
    print(Data)
    print('\n')
```

```python
Grp.get_group('x')
```

```python
Grp.max()
```

```python
Data1.groupby(['Column1']).mean()
```

```python
Data2=Data1.groupby('Column1')['Column2'].agg({'mean'})
Data2
```

```python
Data2.rename(columns={'mean':"Column2"})
```

```python
Dict2={
    'City':['New Delhi','Bangalore','Chennai','Kolkata','Ahemdabad'],
    'Temp':[17,25,27,17,24]
}
```

```python
Dict3={
    'City':['New Delhi','Mumbai','Chennai','Kolkata'],
    'Humidity':[75,77,65,67]
}
```

```python
Data4=pd.DataFrame(Dict3)
Data4
```

```python
# DEFAULT: INNER JOIN (INTERSECTION) 
Data5=pd.merge(Data3,Data4)
Data5
```

```python
# OUTER JOIN (UNION)
Data6=pd.merge(Data3,Data4,how='outer')
Data6
```

```python
# LEFT JOIN: INCLUDE ALL LEFT PART INCLUDING INTERSECTION
Data7=pd.merge(Data3,Data4,how='left')
Data7
```

```python
# RIGHT JOIN: INCLUDE ALL RIGHT PART INCLUDING INTERSECTION
Data8=pd.merge(Data3,Data4,how='right')
Data8
```

```python
Data9=pd.merge(Data3,Data4,how='outer',indicator=True)
Data9
```

```python
Dict4={
    'City':['New Delhi','Bangalore','Chennai','Kolkata','Ahemdabad'],
    'Temp':[17,25,27,17,24],
    'Humidity':[65,57,77,64,74]
}
Data10=pd.DataFrame(Dict4)
Data10
```
