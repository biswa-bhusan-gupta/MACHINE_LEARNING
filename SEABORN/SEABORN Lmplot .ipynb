```python
import seaborn as sns
sns.set_style('darkgrid')
%matplotlib inline
```

```python
x=sns.load_dataset('tips')
x
```

## Lmplot: 

```python
sns.lmplot(x='total_bill',y='tip',data=x)
```

```python
sns.lmplot(x='total_bill',y='tip',data=x,hue='sex')
sns.set_style('ticks')
```

```python
sns.lmplot(x='total_bill',y='tip',data=x,hue='sex',col='day')
sns.set_style('ticks')
sns.despine()
```

```python
sns.lmplot(x='total_bill',y='tip',data=x,row='sex',col='day')
```

## Context:

```python
sns.lmplot(x='total_bill',y='tip',data=x,hue='sex',col='day',aspect=2,height=5)
```
