```python
import seaborn as sns
import matplotlib as plt
%matplotlib inline
sns.set_style('darkgrid')
```

```python
x=sns.load_dataset('iris')
x
```

```python
x.head()
```

## PairGrid:

```python
g=sns.PairGrid(x,palette='coolwarm',hue='species')
g.map(sns.scatterplot)
```

```python
g=sns.PairGrid(x,palette='magma',hue='species')
g.map_diag(sns.distplot,bins=1)
g.map_upper(sns.stripplot)
g.map_lower(sns.swarmplot)
```

```python
sns.pairplot(x,palette='rainbow',hue='species')
```

```python
y=sns.load_dataset('tips')
y
```

## FacetGrid:

```python
h=sns.FacetGrid(y,col='time',row='smoker')
```

```python
h=sns.FacetGrid(y,col='time',row='smoker')
h.map(sns.distplot,'total_bill')
```

```python
h=sns.FacetGrid(y,col='time',row='smoker')
h.map(sns.scatterplot,'total_bill','tip').add_legend()
```

## JointGrid:

```python
o=sns.JointGrid('total_bill','tip',data=y)
o
```

```python
o=sns.JointGrid('total_bill','tip',y)
o.plot(sns.regplot,sns.distplot)
```
