```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set_style('darkgrid')
```

```python
x=sns.load_dataset('tips')
```

```python
x
```

```python
x.head(5)
```

## Barplot:

```python
sns.barplot(x='sex',y='total_bill',data=x,hue='size')
```

```python
import numpy as np
sns.barplot(x='sex',y='total_bill',data=x,estimator=np.std)
```

## Countplot:

```python
sns.set_context('paper')
plt.figure(figsize=(10,6))
sns.countplot(x='sex',data=x)
```

## Boxplot:

```python
sns.boxplot(x='sex',y='total_bill',data=x,hue='size',palette='rainbow')
```

## Violinplot:

```python
sns.violinplot(x='sex',y='total_bill',data=x,hue='time',palette='rainbow')
```

```python
sns.violinplot(x='sex',y='total_bill',data=x,hue='time',palette='Set1')
```

```python
sns.swarmplot(x='sex',y='total_bill',data=x,palette='Set1',hue='time',split=True)
sns.violinplot(x='sex',y='total_bill',data=x,palette='rainbow')
```

## Stripplot:

```python
sns.stripplot(x='sex',y='total_bill',data=x,palette='Set1')
```

## Swarmplot:

```python
sns.swarmplot(x='day',y='total_bill',data=x,palette='rainbow',hue='sex',split=True)
```

## Factorplot:

```python
sns.factorplot(x='sex',y='total_bill',data=x,palette='rainbow',kind='bar')
```
