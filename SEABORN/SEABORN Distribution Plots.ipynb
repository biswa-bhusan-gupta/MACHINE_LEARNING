```python
import seaborn as sns
sns.set_style('darkgrid')
```

```python
%matplotlib inline
```

```python
x=sns.load_dataset('tips')
```

```python
x.head()
```

## Distplot:

```python
sns.distplot(x['total_bill'],kde=False,bins=200)
```

## Jointplot:

```python
sns.jointplot(x='total_bill',y='tip',data=x,kind='scatter')
```

```python
sns.jointplot(x='total_bill',y='tip',data=x,kind='reg')
```

## Pairplot:

```python
sns.pairplot(x,hue='sex',palette='coolwarm')
```

## Rugplot:

```python
sns.rugplot(x['total_bill'])
```

## Kdeplot:

```python
sns.kdeplot(x['total_bill'])
```

```python
sns.rugplot(x['total_bill'])
sns.kdeplot(x['total_bill'])
```
