```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x=np.linspace(0,5,11)
y=x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y,'b*-')
plt.xlabel('X AXIS')
plt.ylabel('Y AXIS')
plt.title('TITLE')
plt.show()
```

```python
plt.subplot(3,3,1)
plt.plot(x,y,'r-*')
plt.subplot(1,2,2)
plt.plot(y,x,'y--')
```

```python
fig=plt.figure()
axes=fig.add_axes([0.2,0.2,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X axis')
axes.set_ylabel('Y axis')
axes.set_title('Economy')
```

```python
fig=plt.figure()
axes1=fig.add_axes([0.2,0.2,0.8,0.8])
axes1.set_xlabel('X Axes')
axes1.set_ylabel('Y Axes')
axes1.set_title('BIGGER PLOT')
axes2=fig.add_axes([0.5,0.3,0.5,0.5])
axes2.set_xlabel('x Axes')
axes2.set_ylabel('y Axes')
axes2.set_title('Smaller Plot')
axes1.plot(x,y,'g*-')
axes2.plot(y,x,'b--')
```

```python
fig,axes=plt.subplots()
axes.plot(x,y,'r*-')
axes.set_xlabel('X axis')
axes.set_ylabel('Y axis')
axes.set_title('Title')
fig
```

```python
fig,axes=plt.subplots(nrows=1,ncols=3)
axes[0].plot(x,x+1,x,x**2)
axes[0].axis("Tight")
axes[1].plot(x,x+2,x,x**3)
axes[1].set_xlim(0.4,6,'g-*')
axes[1].set_ylim(0.6,6,'y-')
axes[2].plot(x,x+3,x,x**4)
plt.tight_layout()
```

```python
fig,axes=plt.subplots(nrows=1,ncols=3)
for i in axes:
    i.plot(x,y)
plt.tight_layout()
```

```python
fig=plt.figure(figsize=(8,4),dpi=100)
```

```python
fig,axes=plt.subplots(figsize=(8,3),nrows=1,ncols=2)
axes[0].plot(x,y,'g')
axes[1].plot(y,x,'y')
plt.tight_layout()
```

```python
fig.savefig('Filename.png',dpi=200)
```

```python
fig=plt.figure()
ax=fig.add_axes([0.2,0.2,0.8,0.8])
ax.plot(x,x**2,'g',label='x**2')
ax.plot(x,x**3,'y',label='x**3')
ax.legend(loc=7)
```

```python
fig,ax=plt.subplots()
ax.plot(x,x+1,color='green',linewidth=2,alpha=2,linestyle='-',marker='+')
ax.plot(x,x+2,color='purple',linewidth=0.5,linestyle='-',marker='o',markersize='10',markerfacecolor='blue',markeredgewidth=2,markeredgecolor='pink')
ax.plot(x,x+3,color='orange',marker='o')
```
